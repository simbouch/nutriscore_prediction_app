Afficher les Chunks par tranche de 10000

```python
import pandas as pd

chunks = pd.read_csv('en.openfoodfacts.org.products.csv',encoding="UTF-8", sep="\t",on_bad_lines='skip', chunksize=10000, low_memory=False)


displayed_lines = []

for chunk in chunks:
    displayed_lines.append(chunk)

len(displayed_lines)

chunk.head(10)
```
